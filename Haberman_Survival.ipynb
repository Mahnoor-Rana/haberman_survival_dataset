## cdf (cumulative distribution function)


```python
# perform univaraite
counts ,bin_edges= np.histogram(haber['age'],bins = 10,density = True)
pdf=counts/sum(counts)
print(pdf)
```

```python
print(bin_edges)
```

```python
#compute cdf
cdf=np.cumsum(pdf)
plt.plot(pdf)
plt.plot(bin_edges)
plt.grid()
plt.show()
```

## BOX PLOT

```python
sns.boxplot(x='year',y='age',data = haber)
plt.show
```

## Violin plot

```python
sns.violinplot(x='year',y = 'age',data=haber,size =4)
plt.show()
```

## bivariate analysis, pair plots,2 D scatter


```python
plt.close();
sns.set_style('whitegrid')
sns.pairplot(haber,hue='year',size = 3)
plt.show()
```

## Scatter plot 


```python
sns.set_style('whitegrid')
sns.FacetGrid(haber,hue='year',data = haber,size= 5).map(plt.scatter,'year','age').add_legend();
plt.show()
```
